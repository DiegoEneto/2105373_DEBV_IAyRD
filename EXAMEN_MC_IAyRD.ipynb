1.- IMPORTAR LAS LIBRERIAS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

2.- INGRESAR EL DATASET Y GUARDARLO

```python
url ="https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv"
df = pd.read_csv(url)
```

3.- PRIMERAS 5 FILAS Y HACER UN RESUMEN ESTADISTICO DESCRIPTIVO

```python
df.head()
df.info()
df.describe()
```

4.- ELIMINAR LOS DATOS NULOS DEL DATAFRAME

```python
df = df[(df[["x","y","z"]] != 0).all(axis=1)]
```

5.- HACER UNA LISTA LLAMADA VOLUMEN_APROX QUE MULTIPLIQUE EL VALOR DE X, Y y Z

```python
df["volumen_aprox"] = df["x"] * df["y"] * df["z"]
df.info()
```

GRAFICAS USANDO MATPLOTLIB

1.- GRAFICO DE CAJA, MUESTRA LA DISTRIBUCION DESCRIPTIVA DE LA COLUMNA PRECIO

```python
plt.boxplot(df["price"])
plt.title("PRECIO",color= "blue")
plt.ylabel("PRECIO",color= "purple")
plt.show()
```

2.- GRAFICO DE DISPERSION DEL VOLUMEN_APROX

```python
plt.scatter(df["volumen_aprox"], df["price"])
plt.title("Volumen aproximado VS Precio", color="blue")
plt.xlabel("Volumen", color= "green")
plt.ylabel("Precio", color= "purple")
plt.show()
```

3.- GRAFICA DE BARRAS RESPECTO A LA CANTIDAD DE PIEZAS PARA CADA EQUIPO

```python
conteo_corte = df["cut"].value_counts()

plt.bar(conteo_corte.index, conteo_corte.values)
plt.title("CANTIDAD DE DIAMANTES POR CADA CORTE", color= "blue")
plt.xlabel("Tipo de corte", color= "green")
plt.ylabel("Cantidad", color= "purple")
plt.show()
```
